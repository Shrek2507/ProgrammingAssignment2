```python
# Here we will practice problems on Joins in Pandas....
import numpy as np
import pandas as pd

class Retail:
    def __init__(self):
        self.salesman_data, self.orders_data, self.customer_data = None,None,None
    
    def read_salesman_data(self):
        # Read Salesman data...
        salesman_data = pd.read_csv('C://Users//Shrey//Desktop//Python_Programs//Pandas_Practice//Salesman.csv')
        
        # Assigning to class variable...
        self.salesman_data = salesman_data
        return self.salesman_data
    
    def read_orders_data(self):
        # Read Orders Data....
        orders_data = pd.read_csv('C://Users//Shrey//Desktop//Python_Programs//Pandas_Practice//Orders.csv')
        
        # Assigning to class variable...
        self.orders_data = orders_data
        return self.orders_data
        
    def read_customer_data(self):
        # Read Customer Data...
        customer_data = pd.read_csv('C://Users//Shrey//Desktop//Python_Programs//Pandas_Practice//Customer.csv')
        
        self.customer_data = customer_data
        return self.customer_data
    
```

```python
# Creating an Object of Retail....
retail = Retail()

# Read salesman and Customer Data...
salesman_data = retail.read_salesman_data()
customer_data = retail.read_customer_data()
```

```python
# prepare a list with salesman name, customer name and their cities for the salesmen and customer who belongs to the same city.
# Perform Inner Join on Salesman_Data and Customer_Data...
print "The Salesman Name, Customer Name and Cities are : "
merged_data = pd.merge(customer_data, salesman_data, how = 'inner', left_on = 'city', right_on = 'city')
merged_data[['name', 'cust_name', 'city']]
```

```python
# make a list with order no, purchase amount, customer name and their cities for those orders which 
# order amount between 500 and 2000.
orders_data = retail.read_orders_data() # Read Orders Data....
orders_data
```

```python
customer_data = retail.read_customer_data()
customer_data
```

```python
# Performing Inner Join between Orders and Customers....
merged_data = pd.merge(orders_data, customer_data, how='inner', left_on = 'customer_id', right_on = 'customer_id')
merged_data
```

```python
merged_data = merged_data[['ord_no', 'purch_amt', 'cust_name', 'city']]
merged_data = merged_data[merged_data['purch_amt'].between(500, 2000)]
merged_data
```

```python
# which salesman are working for which customer.
merged_data = pd.merge(customer_data, salesman_data, how='inner', left_on='salesman_id', right_on = 'salesman_id')
merged_data = merged_data[['cust_name', 'city_x', 'name', 'commission']]
merged_data.columns = ['Customer Name', 'City', 'Salesman Name', 'Commission']
merged_data
```

```python
#  find the list of customers who appointed a salesman for their jobs who gets a commission from the company is more than 12%.
merged_data = pd.merge(customer_data, salesman_data, how='inner', left_on='salesman_id', right_on='salesman_id')
merged_data = merged_data[['cust_name', 'name', 'city_x', 'commission']]
merged_data = merged_data[merged_data['commission'] > 0.12]
merged_data.columns = ['Customer Name', 'Salesman Name', 'City', 'Commission']
merged_data
```

```python
# find the list of customers who appointed a salesman for their jobs who does not live in the same city 
# where their customer lives, and gets a commission is above 12%.
merged_data = pd.merge(customer_data, salesman_data, how = 'inner', left_on = 'salesman_id', right_on ='salesman_id')
merged_data = merged_data[(merged_data['city_x']) <> (merged_data['city_y'])]
merged_data = merged_data[merged_data['commission'] > 0.12]
merged_data = merged_data[['cust_name', 'name', 'city_x', 'city_y', 'commission']]

# Renaming the Columns...
merged_data.columns = ['Customer Name', 'Salesman Name', 'Customer City', 'Salesman City', 'Commission']
merged_data
```
